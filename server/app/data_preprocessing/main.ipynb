```python
import pandas as pd
import numpy as np
import seaborn as sns
import helpers
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
# Drop duplicates
df = helpers.drop_duplicates(df)
# Drop columns with high missing values
df = helpers.drop_columns_with_high_missing_values(df)
# Replace nan values
df = helpers.replace_null_values(df,'embark_town','most_frequent')
df = helpers.replace_null_values(df,'age','most_frequent')
df = helpers.replace_null_values(df,'embarked','most_frequent')
# 
df = helpers.encode(df,'sex','survived')
df.head()
```

```python
df['sex'] = df['sex'].astype(int)
df.dtypes
```
